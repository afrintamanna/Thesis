```python
import pandas as pd 
import numpy as np 
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def readingfiles(filename):
   
    station_df = pd.read_fwf(filename, colspecs='infer', widths=None,skiprows=(np.arange(0,18,1)),encoding="latin1",na_values = "NAN",low_memory=True)
    return(station_df)
```

```python
file_name = "varta nimh 1 at 1.239v 100% discharged.txt"
ni1239v = file_name
ni1239v = readingfiles(ni1239v)

file_name = "varta nimh 1 at 1.276v 75% discharged.txt"
ni1276v = file_name
ni1276v = readingfiles(ni1276v)

file_name = "varta nimh 1 at 1.304v 50% discharged.txt"
ni1304v = file_name
ni1304v = readingfiles(ni1304v)

file_name = "varta nimh 1 at 1.332v 25% discharged.txt"
ni1332v = file_name
ni1332v = readingfiles(ni1332v)

file_name = "varta nimh 1 at 1.364v.txt"
ni1364v = file_name
ni1364v  = readingfiles(ni1364v )

file_name = "varta nimh 2 at 1.265v 100% discharged.txt"
ni1265v = file_name
ni1265v = readingfiles(ni1265v)

file_name = "varta nimh 2 at 1.289v 75% discharged.txt"
ni1289v = file_name
ni1289v  = readingfiles(ni1289v )

file_name = "varta nimh 2 at 1.307v 50% discharged.txt"
ni1307v = file_name
ni1307v = readingfiles(ni1307v)

file_name = "varta nimh 2 at 1.328v 25% discharged.txt"
ni1328v = file_name
ni1328v  = readingfiles(ni1328v )

file_name = "varta nimh 2 at 1.367v.txt"
ni1367v = file_name
ni1367v = readingfiles(ni1367v) 
```

```python
def setting(file): 
    file.index = file["Frequency/Hz"]
    c = file["impedance/Ohm"]
    return(c)
varta1_1239v = setting(ni1239v)
varta1_1276v = setting(ni1276v)
varta1_1304v= setting(ni1304v)
varta1_1332v = setting(ni1332v)
varta1_1364v = setting(ni1364v)
varta2_1265v = setting(ni1265v)
varta2_1289v = setting(ni1289v)
varta2_1307v = setting(ni1307v)
varta2_1328v = setting(ni1328v)
varta2_1367v = setting(ni1367v)
```

```python
fig1, ax1 = plt.subplots(figsize = (10,4), dpi = 600)

ni1239v.plot(ax = ax1, y='impedance/Ohm',x='Frequency/Hz',c="red",marker='x',linewidth=0.2 ,markersize = 5,label= "1239mV")
ni1276v.plot(ax = ax1, y='impedance/Ohm',x='Frequency/Hz',c="black",marker='x',linewidth=0.2 ,markersize = 5,label= "1276mV")
ni1304v.plot(ax = ax1, y='impedance/Ohm',x='Frequency/Hz',c="blue",marker='x',linewidth=0.2 ,markersize = 5,label= "1304mV")
ni1332v.plot(ax = ax1, y='impedance/Ohm',x='Frequency/Hz',c="green",marker='x',linewidth=0.2 ,markersize = 5,label= "1332mV")
ni1364v.plot(ax = ax1, y='impedance/Ohm',x='Frequency/Hz',c="mediumvioletred",marker='x',linewidth=0.2 ,markersize = 5,label= "1364mV")


ax1.grid(True)
ax1.set_xscale('log')
#ax1.set_title("Impedance vs Frequency of VARTA NiMH 1")
plt.ylabel("Impedance/Ohm")
plt.xlabel("Frequency/Hz")
plt.xlim([10**0, 10**5])
plt.savefig("vaarta1.jpeg")
plt.show() 
```
