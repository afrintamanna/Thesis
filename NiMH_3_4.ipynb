```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def readingfiles(filename):
   
    station_df = pd.read_fwf(filename, colspecs='infer', widths=None,skiprows=(np.arange(0,18,1)),encoding="latin1",na_values = "NAN",low_memory=True)
    return(station_df)
```

```python
file_name = "varta nimh 3 at 1.267v 100% discharged.txt"
ni1267v = file_name
ni1267v = readingfiles(ni1267v)

file_name = "varta nimh 3 at 1.280v 75% discharged.txt"
ni1280v = file_name
ni1280v = readingfiles(ni1280v)

file_name = "varta nimh 3 at 1.294v 50% discharged.txt"
ni1294v = file_name
ni1294v = readingfiles(ni1294v)

file_name = "varta nimh 3 at 1.316v 25% discharged.txt"
ni1316v = file_name
ni1316v = readingfiles(ni1316v)

file_name = "varta nimh 3 at 1.353v.txt"
ni1353v = file_name
ni1353v  = readingfiles(ni1353v )

file_name = "varta nimh 4 at 1.307v 50% discharged.txt"
ni1307v = file_name
ni1307v = readingfiles(ni1307v)

file_name = "varta nimh 4 at 1.328v 25% discharged.txt"
ni1328v = file_name
ni1328v  = readingfiles(ni1328v )

file_name = "varta nimh 4 at 1.360v.txt"
ni1360v = file_name
ni1360v = readingfiles(ni1360v) 
```

```python
def setting(file): 
    file.index = file["Frequency/Hz"]
    c = file["impedance/Ohm"]
    return(c)
varta3_1267v = setting(ni1267v)
varta3_1280v = setting(ni1280v)
varta3_1294v= setting(ni1294v)
varta3_1316v = setting(ni1316v)
varta3_1353v = setting(ni1353v)
varta4_1307v = setting(ni1307v)
varta4_1328v = setting(ni1328v)
varta4_1360v = setting(ni1360v)
```

```python
fig1, ax1 = plt.subplots(figsize = (10,4), dpi = 500)

ni1267v.plot(ax = ax1, y='impedance/Ohm',x='Frequency/Hz',c="yellow",marker='x',linewidth=0.2 ,markersize = 5,label= "1267mV")
ni1280v.plot(ax = ax1, y='impedance/Ohm',x='Frequency/Hz',c="black",marker='x',linewidth=0.2 ,markersize = 5,label= "1280mV")
ni1294v.plot(ax = ax1, y='impedance/Ohm',x='Frequency/Hz',c="blue",marker='x',linewidth=0.2 ,markersize = 5,label= "1294mV")
ni1316v.plot(ax = ax1, y='impedance/Ohm',x='Frequency/Hz',c="mediumvioletred",marker='x',linewidth=0.2 ,markersize = 5,label= "1316mV")
ni1353v.plot(ax = ax1, y='impedance/Ohm',x='Frequency/Hz',c="green",marker='x',linewidth=0.2 ,markersize = 5,label= "1353mV")


ax1.grid(True)
ax1.set_xscale('log')
#ax1.set_title("Impedance vs Frequency of VARTA NiMH 3")
ax1.set_ylabel("Impedance/Ohm")
ax1.set_xlabel("Frequency/Hz")
plt.xlim([10**0, 10**5])
plt.savefig("vaarta3.jpeg")
plt.show() 

```
